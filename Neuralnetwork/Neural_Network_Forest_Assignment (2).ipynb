```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('forestfires.csv')
df
```

```python
df.drop(['monthfeb','monthjan','monthjul','monthjun','monthmar','monthmay','monthnov','monthoct','monthsep','monthapr',	'monthaug',	'monthdec'],inplace=True,axis=1)
```

```python
df.drop(['daymon',	'daytue','daywed','daythu','dayfri','daysat','daysun'],inplace=True,axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
cate = ['month','day']
df[cate] = df[cate].apply(le.fit_transform)
```

```python
df['size_category'].replace({'small':0,'large':1},inplace=True)
df
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
from imblearn.over_sampling import SMOTE
from collections import Counter

sm = SMOTE()
X_train,y_train = sm.fit_resample(X,y)

print(f'Before smoting : {Counter(y)}')
print(f'After smoting : {Counter(y_train)}')
```

```python
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
X = sc_x.fit_transform(X)
Y = sc_x.fit_transform(y.values.reshape(-1,1))
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=123)
```

```python
from keras.models import Sequential
from keras.layers import Dense

# Step 1: Prepare the data
# Assuming you have your training data stored in X_train and corresponding labels in y_train

# Convert data to numpy arrays



# Step 2: Create the model
model = Sequential()
model.add(Dense(12, input_dim=11, activation='relu'))
model.add(Dense(10,  activation='relu')) # Hidden layer with 12 neurons
model.add(Dense(1, activation='sigmoid'))  # Output layer with sigmoid activation for binary classification

# Step 3: Compile the model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Step 4: Train the model
model.fit(X_train, y_train, epochs=20, batch_size=10)

# Step 5: Use the trained model for prediction
# Assuming you have your test data stored in X_test
predictions = model.predict(X_test)
```
