```python
# to know how many user has read more than 200 books
x =  df['user_id'].value_counts() > 100
y = x[x].index

users = df[df['user_id'].isin(y)]

```

```python
users
```

```python
book_pivot = users.pivot_table(columns='book_title', index='user_id', values="book_rating")
book_pivot.fillna(0, inplace=True)
book_pivot
```

```python
book_pivot.columns
```

```python
user_rating = book_pivot['11 Edward Street']
```

```python
similar_to_edward = book_pivot.corrwith(user_rating)
similar_to_edward
recommended = pd.DataFrame(similar_to_edward,columns = ['correlation'])
```

```python
recommended.sort_values('correlation',ascending=False).head(30)
```

```python
book_pivot1 = users.pivot_table(columns='user_id', index='book_title', values="book_rating")
book_pivot1.fillna(0, inplace=True)
# book_pivot1.index
```

```python
book_pivot1.index[book_pivot1.index=='11 Edward Street'].tolist()
```

```python
from scipy.sparse import csr_matrix
book_sparse = csr_matrix(book_pivot1)
```

```python
from sklearn.neighbors import NearestNeighbors

classifier =NearestNeighbors(n_neighbors=5,algorithm='brute')
classifier.fit(book_sparse)
```

```python
dis,recommend_book = classifier.kneighbors(book_pivot1.loc['11 Edward Street',:].values.reshape(1,-1))
```

```python
for i in range(len(recommend_book)):
  print(book_pivot1.index[recommend_book[i]])
```
